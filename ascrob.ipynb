```python
#coding:utf-8
Dict = {}
def printListenersTimesPlayed(userID):
    with open("user_artist_data.txt", 'r', encoding='utf-8') as file:
        for line in file:
            data = line.rstrip().split()
            if data[0] == str(userID):
                Dict.update({data[1]: data[2]})
    
    with open("artist_data.txt", 'r', encoding='utf-8') as file:
        for line in file:
            data2 = line.rstrip().split()
            if data2: 
                if data2[0] in Dict:
                    print(data2[1] + '-' + Dict[data2[0]])

printListenersTimesPlayed(1000019)                
```
